채점해볼 수 있는 페이크 데이터 생성

```python
# 지명 데이터 생성
# 지명 문항은 PN1-PN19까지 총 19문항
# 지명 문항 데이터 불러오기
import pandas as pd
import requests
from io import BytesIO
url = "https://github.com/Byumin/peer/raw/refs/heads/item/item_info.xlsx"
response = requests.get(url)
response.raise_for_status()
df = pd.read_excel(BytesIO(response.content))
df
point_item = df[df['reporting']=='또래지명']
point_item
```

```python
# 자기보고 문항 데이터 불러오기
self_item = df[df['reporting']=='자기보고(5점 리커트)']
self_item
```

```python
# 20명의 학생의 19문항의 응답 데이터 생성
# 1대1 서로가 서로를 랜덤하게 지목한 데이터 생성
import numpy as np
import pandas as pd
import random

# 학생 수 및 문항 수 정의
num_students = 20
num_questions = len(point_item)

# 학생 이름 생성
students = [f"student{i+1}" for i in range(num_students)]

# 1대 다수 또래지명 데이터 생성_동일한 학생을 중복 지목하지 않음
point_df = pd.DataFrame()

for i in range(num_questions) :
    np.random.seed(i)
    #pairings = np.random.permutation(students)
    point_item_no = [f'PN{i+1}']*num_students
    print(point_item_no)
    student_designation = students.copy()
    
    #가중치 생성하기 위해 난수
    rand_values = [x for x in np.round(np.random.rand(len(students)),2)]
    weights = [x/sum(rand_values) for x in rand_values]
    print(weights)
    
    student_nomination = random.choices(students, weights=weights, k=num_students)
    #student_nomination = random.sample(students, num_students)
    temp_df = pd.DataFrame({'point_item_no' : point_item_no, 'student_designation' : student_designation, 'student_nomination' : student_nomination})
    point_df = pd.concat([point_df, temp_df], axis=0, ignore_index=True)
    
point_df
```

```python
# 제대로 나왔는지 확인
point_df.to_excel('point_df.xlsx', index=False)
```

```python
# 자기보고 응답 데이터 생성
# 10명의 학생

num_students = 20
num_questions = len(self_item)

# 학생 이름 생성
students = [f"student{i+1}" for i in range(num_students)]

# 문항 응답 데이터 생성 (1~5점 랜덤 부여)
responses = np.random.randint(1, 6, size=(num_students, num_questions))
responses = pd.DataFrame(responses, columns=self_item['no'].to_list())

students_df = pd.DataFrame({'students' : students})
self_df = pd.concat([students_df, responses], axis=1, ignore_index=False)
self_df
```

```python
# 제대로 나왔는지 확인
self_df.to_excel('self_df.xlsx', index=False)
```

```python
# 1 대 다수 지명 데이터 생성
# 20명의 학생의 19문항의 응답 데이터 생성
import numpy as np
import pandas as pd
import random

# 학생 수 및 문항 수 정의
num_students = 20
num_questions = len(point_item)

# 학생 이름 생성
students = [f"student{num+1}" for num in range(num_students)]

# 1대 다수 또래지명 데이터 생성_동일한 학생을 중복 지목하지 않음
point_df = pd.DataFrame()

for i in range(num_questions) :
    df = pd.DataFrame()
    for student in students :
        # 학생이 지목할 명수 생성, 정규분포 활용해 (k<학생수), 1 이상 지명
        #k = max(1, min(int(np.random.normal(loc=1, scale=1, size=1)[0]), num_students-1))
        k = max(1, np.random.poisson(lam=1, size=1)[0])
        
        # 학생이 k명을 본인을 제외한 학생(n-1) 중 지목할 변수 (k<n-1)
        able_students = [stu for stu in students if stu!=student]
        rand_values = [x for x in np.round(np.random.rand(len(able_students)),2)]
        weights = [x/sum(rand_values) for x in rand_values] # 합이 1인
        print(k)
        student_nomination = np.random.choice(able_students, p=weights, size=k, replace=False).tolist()

        temp_df = pd.DataFrame({'point_item_no' : [f'PN{i+1}']*k, 'student_designation' : [student]*k, 'student_nomination' : student_nomination})
        df = pd.concat([df, temp_df], axis=0, ignore_index=True)
    point_df = pd.concat([point_df, df], axis=0, ignore_index=True)
point_df
```

```python
# 제대로 나왔는지 확인
point_df.to_excel('point_df.xlsx', index=False)
```
