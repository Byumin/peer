### 공격성 규범

```python
import pandas as pd

df = pd.read_excel('5학년_1반_전수(AC000120255019FA1)/AC000120255019FA1_또래지명_응답.xlsx')
print(df)
```

```python
# 학생 명단 먼저 확보

student_df = pd.read_excel('5학년_1반_전수(AC000120255019FA1)/AC000120255019FA1_학생 명단.xlsx')
student_list = student_df['학생 이름'].to_list()
print(student_list)
```

```python
df.dtypes
```

```python
df['문항번호'].isin([4,5,6,7])
```

```python
aggression_item_no = [4,5,6,7]
temp_df=df[df['문항번호'].isin(aggression_item_no)]
print(temp_df)
```

```python
# 지명수 구하는 함수
def count_point(df, student_list, item_no) :
    temp_df = df[df['문항번호'].isin(item_no)]
    count_list = []
    for student in student_list :
        if len(temp_df[temp_df['지명받은 학생']==student]) >= 1 :
            freq = len(temp_df[temp_df['지명받은 학생']==student])
        elif len(temp_df[temp_df['지명받은 학생']==student]) == 0 :
            freq = 0
        count_list.append(freq)
    count_point_df = pd.DataFrame({'학생 이름': student_list, '지명수': count_list})
    return count_point_df
```

```python
count_point_df = count_point(df, student_list, aggression_item_no)
print(count_point_df)
```

```python
'''
--- !!! ---
25.06.30 오후 3시 37분
공격성 규범은 학생(윤지후, 양지호, 조수아, 조서윤) 개별 응답과 산출값 간 불일치로 인해
학급 분위기 결과 역시 정확한 검증이 어려움.
이에 따라 해당 항목의 결과값 검수는 보류함.
'''
```

```python
# 함수는 만들어 놓고
aggression_item_no = [4,5,6,7]
item_no14 = [14]
aggression_count_point_df = count_point(df, student_list, aggression_item_no)
no14_count_point_df = count_point(df, student_list, item_no14)
print(aggression_count_point_df)
print(no14_count_point_df)
multiply_df = aggression_count_point_df.merge(no14_count_point_df, on='학생 이름')
multiply_df.columns = ['학생 이름', '공격지명수', '친구지명수']
multiply_df['공격지명*친구 지명'] = multiply_df['공격지명수'] * multiply_df['친구지명수']
print(multiply_df)
multiply_df.sort_values(by='공격지명*친구 지명', ascending=False, inplace=True)
print(multiply_df)
top5_df = multiply_df.head(5)
top5_mean = top5_df['공격지명*친구 지명'].mean()
total_mean = multiply_df['공격지명*친구 지명'].mean()
weighted_mean = 0.4 * (top5_mean) + 0.6 * (total_mean)
print(weighted_mean)
weighted_mean = 4 * ((min(max(weighted_mean, 0), 67)-0)/67)+1
print(top5_mean, total_mean)
print(f'공격성 규범 : {weighted_mean}')
```

```python
'''
--- !!! ---
25.06.30 오후 5시 00분
공격성 규범은 학생(윤지후, 양지호, 조수아, 조서윤) 개별 응답과 산출값 간 불일치인데
학급 분위기 결과는 일치함.
'''
# ! 가중평균의 파라미터 설정을 다시 해야할 수 있음. 5학년 1반_전수(AC000120255019FA1)에서 공격성 규범이 충분히 커보이는데 낮게 결과값이 낮게 나타남
```

### 지위위계

```python
import pandas as pd

df = pd.read_excel('5학년_1반_전수(AC000120255019FA1)/AC000120255019FA1_또래지명_응답.xlsx')
print(df)
```

```python
# 학생 명단 먼저 확보

student_df = pd.read_excel('5학년_1반_전수(AC000120255019FA1)/AC000120255019FA1_학생 명단.xlsx')
student_list = student_df['학생 이름'].to_list()
print(student_list)
```

```python
df.dtypes
```

```python
item_no = [1]
temp_df=df[df['문항번호'].isin(item_no)]
print(temp_df)
```

```python
# 지명수 구하는 함수
def count_point(df, student_list, item_no) :
    temp_df = df[df['문항번호'].isin(item_no)]
    count_list = []
    for student in student_list :
        if len(temp_df[temp_df['지명받은 학생']==student]) >= 1 :
            freq = len(temp_df[temp_df['지명받은 학생']==student])
        elif len(temp_df[temp_df['지명받은 학생']==student]) == 0 :
            freq = 0
        count_list.append(freq)
    count_point_df = pd.DataFrame({'학생 이름': student_list, '지명수': count_list})
    return count_point_df
```

```python
'''
--- !!! ---
25.06.30 오후 3시 41분
지위위계는 학생 개별 응답과 산출값 일치하여
학급 분위기 결과 검증 진행중.
'''
```

```python
def gini(count_point_df) :
    point_mean = count_point_df['지명수'].mean()
    print(f'지명수 평균 : {point_mean}')
    n = len(count_point_df)
    print(f'학생 수 : {n}')
    abs_diff_list = []
    for point in count_point_df['지명수'] :
        abs_diff = abs(point - point_mean)
        abs_diff_list.append(abs_diff)
    abs_diff_sum = sum(abs_diff_list)
    print(f'절대차 합계 : {abs_diff_sum}')
    gini = abs_diff_sum / (2 * n * point_mean)
    print(f'지위위계 Gini 계수 : {gini}')
    gini_normalization = 4 * ((min(max(gini, 0), 0.6) - 0)/ 0.6) +1
    return gini_normalization

gini_normalization = gini(count_point_df)
print(f'학급 분위기_지위위계 : {gini_normalization}')
```

```python
def gini(count_point_df) :
    point_mean = count_point_df['지명수'].mean()
    print(f'지명수 평균 : {point_mean}')
    n = len(count_point_df)
    print(f'학생 수 : {n}')
    abs_diff_list = []
    for i in count_point_df['지명수'] :
        for j in count_point_df['지명수'] :
            if i != j:
                abs_diff = abs(i - j)
                abs_diff_list.append(abs_diff)
            else :
                pass
    abs_diff_sum = sum(abs_diff_list)
    print(f'절대차 합계 : {abs_diff_sum}')
    gini = abs_diff_sum / (2 * n * n * point_mean)
    print(f'지위위계 Gini 계수 : {gini}')
    gini_normalization = 4 * ((min(max(gini, 0), 0.6) - 0)/ 0.6) +1
    return gini_normalization

gini_normalization = gini(count_point_df)
print(f'학급 분위기_지위위계 : {gini_normalization}')
```
