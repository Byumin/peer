# 목적 : 스쿨프렌즈 검수 데이터 생성

## 내용  
GPT에서 아래와 같은 자료 생성  
* 역할과 척도에 따른 가중치 맵핑 자료  
* 학생 목록  
* 학생의 역할

필요한 소스  
1. 역할에 따른 가중치를 확률적으로 고려한 지명 데이터
2. 역할에 따른 가중치를 확률적으로 고려한 1~5점 데이터

```python
# 역할에 따른 가중치를 확률적으로 고려한 지명 데이터
# 학생 명단 불러오기
import pandas as pd
students_name_rule = pd.read_excel('3차(정서윤 추가, 이서연 제거)/student_role_mapping_updated.xlsx')
```

```python
students_name_rule.columns = ['번호', '학생명', '역할', '활성화_유무']
```

```python
students_name_rule
```

```python
# nan값 활성화로 대체
students_name_rule['활성화_유무'] = students_name_rule['활성화_유무'].fillna('활성화')
students_name_rule
```

```python
# 역할에 따른 가중치 데이터 불러오기
role_weight = pd.read_excel('3차(정서윤 추가, 이서연 제거)/roles_weight_mapping_final.xlsx')
role_weight
```

```python
# 역할에 따른 가중치와 학생 역할 매핑자료 병합
students_role_weight = pd.merge(students_name_rule, role_weight, how='left', left_on='역할', right_on='역할')
# 가중치 2.2만  5로 조정
weights_col = students_role_weight.columns[4:]
for col in weights_col:
    students_role_weight[col] = students_role_weight[col].replace(2.2, 5)
    students_role_weight[col] = students_role_weight[col].replace(0.5, 0.1)
    students_role_weight[col] = students_role_weight[col].replace(1, 0.6)
students_role_weight
```

```python
# 지명 문항 정보 불러오기
point_question_info = pd.read_excel('3차(정서윤 추가, 이서연 제거)/250825_스쿨프렌즈 검사 최종 문항_저자수정요청.xlsx', sheet_name='학생문항_또래지명')
point_question_info
```

```python
point_question_info['scale']
```

```python
# point_question_info의 scale에 해당하는 시리즈를 컬럼으로 전환
# 학생 이름과 롤을 행인덱싱으로 활용
# 틀만 우선 만들고
point_question_simulation_t = students_role_weight[['번호','학생명','역할']].copy()
for i, new_col in enumerate(point_question_info['scale']):
    point_question_simulation_t.loc[:, f'{i}_{new_col}'] = pd.NA
point_question_simulation_t.head()
```

```python
# 어떻게 지명 시뮬레이션을 하지~??
# 문항 척도가 컬럼명이고
# 가중치 변수가 척도 이름이니 이 두개가 일치할 경우
# 학생별로 가중치에 따라 확률적으로 지명하는 방식
# 예를 들어 '지각된 인기' 문항 척도인 경우 각 학생의 '지각된 인기' 가중치를 가지고 오고
# 이 가중치를 기반으로 확률적으로 지명해 학생 이름을 산출
# 예시로 하나만 만들어 볼까
import numpy as np
students_name_list = list(point_question_simulation_t['학생명'])
# 본인 제거
students_name_list.remove('김민준')
weights = list(students_role_weight.loc[students_role_weight['학생명'] != '김민준', '지각된 인기'])
students_weights_dict = dict(zip(students_name_list, weights))
students_weights_dict
n = np.random.choice([1,2,3,4,5], p=[0.3, 0.4, 0.2, 0.05, 0.05])  # 1~5명을 확률적으로 지명
point = np.random.choice(students_name_list, p=np.array(weights)/sum(weights), replace=False, size=n)
print(point)
point = ", ".join(point.tolist())
print(point)
```

```python
students_name_list
```

```python
# for문으로 지명 데이터 생성
students_name_list = list(point_question_simulation_t['학생명'])
students_name_list.remove('이서연')  # 이서연 비활성화

# 비활성화한 학생 제외하는 필터링
students_role_weight = students_role_weight[students_role_weight['학생명'].isin(students_name_list)]

# weights_col 정재
weights_col = point_question_simulation_t.columns[3:]

for name in students_name_list:
    students_name_list_copy = students_name_list.copy()
    print(name)
    students_name_list_copy.remove(name)  # 본인 제거
    for col in weights_col:
        # col의 숫자와 _ 제거
        col_refined = col.split('_', 1)[1]
        weights = list(students_role_weight.loc[students_role_weight['학생명'] != name, col_refined])
        n = np.random.choice([1,2,3,4,5], p=[0.3, 0.4, 0.2, 0.05, 0.05])  # 1~5명을 확률적으로 지명
        point = np.random.choice(students_name_list_copy, p=np.array(weights)/sum(weights), replace=False, size=n)
        point = ", ".join(point.tolist())
        point_question_simulation_t.loc[point_question_simulation_t['학생명'] == name, col] = point
point_question_simulation_t
```

```python
# 시뮬레이션 결과 저장
point_question_simulation_t.to_excel('3차(정서윤 추가, 이서연 제거)/student_point_question_simulation_final.xlsx', index=False)
```

```python
# 자기보고 문항 데이터 불러오기
self_report_question_info = pd.read_excel('3차(정서윤 추가, 이서연 제거)/250825_스쿨프렌즈 검사 최종 문항_저자수정요청.xlsx', sheet_name='학생문항_자기보고')
self_report_question_info
```

```python
# self_report_question_info의 scale에 해당하는 시리즈를 컬럼으로 전환
# 학생 이름과 롤을 행인덱싱으로 활용
# 틀만 우선 만들고
self_report_question_simulation_t = students_role_weight[['번호','학생명','역할']].copy()
for index, new_col in enumerate(self_report_question_info['scale']):
    self_report_question_simulation_t.loc[:, f'{index}_{new_col}'] = pd.NA
self_report_question_simulation_t
```

```python
weights_col = self_report_question_simulation_t.columns[3:]
print(weights_col)
```

```python
students_role_weight.columns
```

```python
weight_low_self_prob = [0.4, 0.3, 0.2, 0.05, 0.05]
weight_mid_self_prob = [0.1, 0.2, 0.4, 0.2, 0.1]
weight_high_self_prob = [0.05, 0.05, 0.2, 0.3, 0.4]
```

```python
for name in students_name_list:
    print(name)
    for col in weights_col:
        # col의 숫자와 _ 제거
        col_refined = col.split('_', 1)[1]
        weight = students_role_weight.loc[students_role_weight['학생명'] == name, col_refined].values[0]
        print(col_refined, weight)
        if weight == 5.0:
            score_prob = weight_high_self_prob
        elif weight == 0.6:
            score_prob = weight_mid_self_prob
        else:
            score_prob = weight_low_self_prob
        base_scores = np.array([1, 2, 3, 4, 5])
        score = np.random.choice(base_scores, p=score_prob)
        self_report_question_simulation_t.loc[self_report_question_simulation_t['학생명'] == name, col] = score
```
