# python으로 아테네 테이블 읽어서 복호화 후 저장

```python
import boto3
import re

# S3 클라이언트 생성
s3 = boto3.client("s3")

bucket_name = "schoolfriends-bym"

# 버킷 내 모든 객체 가져오기
response = s3.list_objects_v2(Bucket=bucket_name)
```

```python
for obj in response.get("Contents", []):
    key = obj["Key"]
    if key.endswith(".csv"):
        print(f"Processing file: {key}")
        s3.download_file(bucket_name, key, key.split("/")[-1])
```
